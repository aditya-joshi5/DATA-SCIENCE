ques 1 What is a Decision Tree, and how does it work?

A decision tree is a visual and analytical tool used to make decisions by mapping out possible outcomes, costs, and benefits. It is widely utilized in various fields, including machine learning, finance, healthcare, and operations.



ques 2 What are impurity measures in Decision Trees?

In decision trees, impurity measures assess how well a dataset is split at each node, guiding the algorithm in selecting the best feature and threshold for partitioning the data. Two commonly used impurity measures are Gini Impurity and Entropy.

* Gini Impurity quantifies the likelihood of a randomly chosen element being incorrectly classified.

* Entropy measures the uncertainty or disorder within a dataset

ques 5 What is Information Gain, and how is it used in Decision Trees?

Information Gain is a crucial metric in decision tree algorithms, guiding the selection of features that best reduce uncertainty about the target variable. By iteratively applying Information Gain, decision trees can effectively partition data to achieve accurate classifications.

ques 6 What is the difference between Gini Impurity and Entropy?

**Gini Impurity**: Focuses on the likelihood of misclassification. A lower Gini value indicates a purer node, meaning most samples belong to a single class.

* Gini Impurity is computationally less expensive since it doesn't involve logarithmic calculations, making it preferable for large datasets.

* 	Faster to compute (no logarithmic calculations).


**Entropy**: Measures the unpredictability or disorder. A higher entropy value suggests a more mixed class distribution, indicating greater uncertainty.

* Empirical studies show that both metrics often lead to similar decision trees, with only minor differences in performance.

* More computationally intensive due to logarithmic operations.



ques 7  What is the mathematical explanation behind Decision Trees ?

decision trees use mathematical criteria like Gini impurity and entropy to recursively partition the data, aiming to create homogeneous subsets that predict the target variable accurately. The process involves selecting the best features and thresholds at each node, building the tree from the root to the leaves, and pruning to improve generalization.

ques 8  What is Pre-Pruning in Decision Trees?

Pre-pruning is a proactive approach to control the growth of a decision tree, aiming to balance model complexity and generalization. While it offers benefits like reduced overfitting and faster training times, careful selection of stopping criteria is crucial to avoid underfitting.

ques 9 What is Post-Pruning in Decision Trees?

Post-pruning is a valuable technique for refining decision trees, especially when the goal is to achieve high accuracy and generalization. While it requires more computational resources and careful tuning, the benefits often outweigh the drawbacks in many real-world applications.

ques 10 What is the difference between Pre-Pruning and Post-Pruning?

**Pre-Pruning** : Halts the growth of the decision tree before it reaches its full depth.

* Potential underfitting if constraints are too strict.

* Lower computational Cost , as tree is not fully grown.

**Post-Pruning** : Allows the decision tree to grow fully and then removes branches that do not provide significant predictive power.

* Potential overfitting if pruning is not done effectively.

* Higher  computational Cost, due to full tree construction and subsequent pruning.

ques 11 What is a Decision Tree Regressor?

A Decision Tree Regressor is a supervised machine learning algorithm used for predicting continuous target variables. It models data by recursively splitting it into subsets based on feature values, forming a tree-like structure where each leaf node represents a predicted continuous value.



ques 12  What are the advantages and disadvantages of Decision Trees?

Advantages of Decision Trees:

* Decision Trees provide a clear, visual representation of decision-making processes, making them accessible even to non-experts.

* They can handle both numerical and categorical data without the need for extensive scaling or normalization

* Decision Trees can manage missing data by learning surrogate splits, reducing the impact of incomplete datasets

Disadvantages of Decision Trees:

* Small changes in the training data can result in significantly different tree structures, making them sensitive to data variations.

* Decision Trees can be unstable, meaning that a small change in the data can lead to a large change in the structure of the optimal decision tree.

* Without proper pruning or constraints, Decision Trees can become overly complex, capturing noise in the data and leading to poor generalization on unseen data.

ques 13  How does a Decision Tree handle missing values ?

Decision Trees are adept at managing missing values through strategies like attribute splitting, weighted impurity calculations, and surrogate splits. These built-in mechanisms allow them to maintain robustness and predictive accuracy even in the presence of incomplete data.


ques 14 How does a Decision Tree handle categorical features?

Handling Categorical Features in Decision Trees:

**One-Hot Encoding**: This method transforms each category into a separate binary feature. For example, a "Color" feature with categories {Red, Green, Blue} becomes three binary features: "IsRed", "IsGreen", and "IsBlue".

**Label Encoding**: Each category is assigned a unique integer label. While this method is straightforward, it can introduce unintended ordinal relationships, which might not be appropriate for nominal data.


**Surrogate Splits**: Some Decision Tree implementations can use surrogate splits, where alternative features are considered when the primary feature for a split has missing values. This ensures that the tree can still make decisions even with incomplete data.

```python

from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score


iris = load_iris()
```

```python
# ques 16  Write a Python program to train a Decision Tree Classifier on the Iris dataset and print the model accuracy.

X = iris.data
y = iris.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
clf = DecisionTreeClassifier(random_state=42)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Model Accuracy: {accuracy * 100:.2f}%")
```

```python
# ques 17  Write a Python program to train a Decision Tree Classifier using Gini Impurity as the criterion and print the feature importances


import matplotlib.pyplot as plt
import seaborn as sns
X = iris.data
y = iris.target
feature_names = iris.feature_names
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
clf = DecisionTreeClassifier(criterion='gini', random_state=42)
clf.fit(X_train, y_train)
importances = clf.feature_importances_
import pandas as pd
importance_df = pd.DataFrame({
    'Feature': feature_names,
    'Importance': importances
}).sort_values(by='Importance', ascending=False)
print(importance_df)
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.barplot(x='Importance', y='Feature', data=importance_df)
plt.title('Feature Importances in Decision Tree Classifier (Gini Impurity)')
plt.show()
```

```python
# ques 18 Write a Python program to train a Decision Tree Classifier using Entropy as the splitting criterion and print the model accuracy

X = iris.data
y = iris.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
clf = DecisionTreeClassifier(criterion='entropy', random_state=42)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Model Accuracy: {accuracy * 100:.2f}%")
```

```python
# ques 23 Write a Python program to apply feature scaling before training a Decision Tree Classifier and compare its accuracy with unscaled data

from sklearn.preprocessing import StandardScaler
X = iris.data
y = iris.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
clf_default = DecisionTreeClassifier(random_state=42)
clf_default.fit(X_train, y_train)
clf_scaled = DecisionTreeClassifier(random_state=42)
clf_scaled.fit(X_train_scaled, y_train)
y_pred_default = clf_default.predict(X_test)
accuracy_default = accuracy_score(y_test, y_pred_default)
y_pred_scaled = clf_scaled.predict(X_test_scaled)
accuracy_scaled = accuracy_score(y_test, y_pred_scaled)
print(f"Accuracy without scaling: {accuracy_default * 100:.2f}%")
print(f"Accuracy with scaling: {accuracy_scaled * 100:.2f}%")
```

```python
# ques 26  Write a Python program to train a Decision Tree Regressor with max_depth=5 and compare its performance with an unrestricted tree

# Instead of load_boston, use fetch_california_housing
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error
housing = fetch_california_housing()
X = housing.data
y = housing.target


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
clf_depth_5 = DecisionTreeRegressor(max_depth=5, random_state=42)
clf_depth_5.fit(X_train, y_train)
clf_full = DecisionTreeRegressor(random_state=42)
clf_full.fit(X_train, y_train)
y_pred_depth_5 = clf_depth_5.predict(X_test)
mse_depth_5 = mean_squared_error(y_test, y_pred_depth_5)
y_pred_full = clf_full.predict(X_test)
mse_full = mean_squared_error(y_test, y_pred_full)
print(f"MSE with max_depth=5: {mse_depth_5:.2f}")
print(f"MSE with fully grown tree: {mse_full:.2f}")
```

```python
# ques 27 Write a Python program to train a Decision Tree Classifier, apply Cost Complexity Pruning (CCP), and visualize its effect on accuracy

from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt
X = iris.data
y = iris.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
clf = DecisionTreeClassifier(random_state=42)
clf.fit(X_train, y_train)
accuracy_unpruned = clf.score(X_test, y_test)
print(f"Accuracy of unpruned tree: {accuracy_unpruned * 100:.2f}%")
path = clf.cost_complexity_pruning_path(X_train, y_train)
ccp_alphas, impurities = path.ccp_alphas, path.impurities
pruned_trees = []
for ccp_alpha in ccp_alphas:
    pruned_clf = DecisionTreeClassifier(random_state=42, ccp_alpha=ccp_alpha)
    pruned_clf.fit(X_train, y_train)
    pruned_trees.append(pruned_clf)
accuracies_pruned = [tree.score(X_test, y_test) for tree in pruned_trees]
best_alpha = ccp_alphas[accuracies_pruned.index(max(accuracies_pruned))]
best_pruned_tree = pruned_trees[accuracies_pruned.index(max(accuracies_pruned))]
accuracy_pruned = best_pruned_tree.score(X_test, y_test)
print(f"Accuracy of best pruned tree (alpha={best_alpha}): {accuracy_pruned * 100:.2f}%")
plt.figure(figsize=(12, 8))
plot_tree(clf, filled=True, feature_names=iris.feature_names, class_names=iris.target_names)
plt.title("Unpruned Decision Tree")
plt.show()
plt.figure(figsize=(12, 8))
plot_tree(best_pruned_tree, filled=True, feature_names=iris.feature_names, class_names=iris.target_names)
plt.title(f"Pruned Decision Tree (alpha={best_alpha})")
plt.show()
```

```python
# ques 28 Write a Python program to train a Decision Tree Classifier and evaluate its performance using Precision, Recall, and F1-Score

from sklearn.metrics import precision_score, recall_score, f1_score, classification_report

X = iris.data
y = iris.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
clf = DecisionTreeClassifier(random_state=42)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
precision = precision_score(y_test, y_pred, average='weighted')
recall = recall_score(y_test, y_pred, average='weighted')
f1 = f1_score(y_test, y_pred, average='weighted')

print(f"Precision: {precision * 100:.2f}%")
print(f"Recall: {recall * 100:.2f}%")
print(f"F1-Score: {f1 * 100:.2f}%")
print("\nClassification Report:")
print(classification_report(y_test, y_pred, target_names=iris.target_names))
```

```python
# ques 29 Write a Python program to train a Decision Tree Classifier and visualize the confusion matrix using seaborn

from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

X = iris.data
y = iris.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
clf = DecisionTreeClassifier(random_state=42)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=iris.target_names, yticklabels=iris.target_names)
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Confusion Matrix for Decision Tree Classifier')
plt.show()
```

```python
# ques 30 Write a Python program to train a Decision Tree Classifier and use GridSearchCV to find the optimal values for max_depth and min_samples_split.

from sklearn.model_selection import train_test_split, GridSearchCV
X = iris.data
y = iris.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
dt = DecisionTreeClassifier(random_state=42)
param_grid = {
    'max_depth': [3, 5, 10, None],
    'min_samples_split': [2, 5, 10]
}
grid_search = GridSearchCV(estimator=dt, param_grid=param_grid, cv=5, scoring='accuracy')
grid_search.fit(X_train, y_train)
print("Best Hyperparameters:", grid_search.best_params_)
print("Best Cross-Validation Accuracy:", grid_search.best_score_)
best_model = grid_search.best_estimator_
test_accuracy = best_model.score(X_test, y_test)
print("Test Accuracy of Best Model:", test_accuracy)
```
