**Theoretical**

ques 1  What is a Support Vector Machine (SVM)?

A Support Vector Machine (SVM) is a supervised machine learning algorithm used for classification and regression tasks. It works by finding the optimal hyperplane that best separates data points of different classes in a high-dimensional space.

ques 2 What is the difference between Hard Margin and Soft Margin SVM ?

Hard Margin SVM

Assumption: The data is perfectly linearly separable, meaning there exists a clear boundary that separates the classes without any overlap.

Objective: Maximize the margin between the classes while ensuring that all data points are correctly classified.

Use Case: Suitable for clean datasets without noise or overlapping classes.

 Soft Margin SVM

Assumption: The data may not be perfectly separable, allowing for some misclassifications.

Objective: Balance between maximizing the margin and minimizing misclassification errors.

Use Case: Ideal for real-world datasets with noise, outliers, or overlapping classes.

ques 3  What is the mathematical intuition behind SVM ?

The mathematical intuition behind SVMs involves:

* Maximizing the margin: Finding the hyperplane that maximizes the distance between the classes.

* Solving an optimization problem: Formulating the problem as a convex optimization task to find the optimal hyperplane.

* Utilizing support vectors: Identifying the critical data points that define the hyperplane.

* Applying the kernel trick: Mapping data to higher-dimensional spaces to handle non-linear separations

ques 4  What is the role of Lagrange Multipliers in SVM ?

Lagrange multipliers in SVMs serve to incorporate the constraints into the optimization process, facilitating the derivation of the dual problem. This approach not only simplifies the computation but also highlights the critical role of support vectors in defining the optimal hyperplane

ques 5  What are Support Vectors in SVM?

Support vectors are the essential data points that lie closest to the decision boundary in SVMs. They play a crucial role in defining the optimal hyperplane and contribute significantly to the model's performance and generalization ability

ques 6 What is a Support Vector Classifier (SVC)?

The Support Vector Classifier is a powerful tool for classification tasks, capable of handling both linear and non-linear data. By focusing on support vectors and maximizing the margin, SVC aims to create a robust model that generalizes well to unseen data

ques 7  What is a Support Vector Regressor (SVR)?

Support Vector Regression is a powerful and flexible regression model that can handle both linear and non-linear relationships. Its robustness to outliers and effectiveness in high-dimensional spaces make it a valuable tool for various predictive modeling tasks.

ques 8 What is the Kernel Trick in SVM ?

The Kernel Trick is a pivotal concept in Support Vector Machines (SVMs) that allows them to efficiently handle non-linearly separable data by implicitly mapping it into a higher-dimensional space where a linear separation is possible.

ques 9 Compare Linear Kernel, Polynomial Kernel, and RBF Kernel .

Linear Kernel: Best for linearly separable data and high-dimensional sparse datasets.

Polynomial Kernel: Effective for data with polynomial relationships; requires careful tuning of parameters.

RBF Kernel: Highly flexible and suitable for complex, non-linear data; sensitive to parameter choices

ques 10 What is the effect of the C parameter in SVM ?

The C parameter in Support Vector Machines (SVMs) is a crucial hyperparameter that governs the trade-off between achieving a low error on the training data and maintaining a model that generalizes well to unseen data. Effectively, it controls the regularization of the model


ques 11  What is the role of the Gamma parameter in RBF Kernel SVM ?

The gamma (γ) parameter in the Radial Basis Function (RBF) kernel of Support Vector Machines (SVMs) plays a pivotal role in determining the model's complexity and its ability to generalize to unseen data.

ques 12  What is the Naïve Bayes classifier, and why is it called "Naïve" ?

The Naïve Bayes classifier is a family of probabilistic algorithms based on Bayes' Theorem, commonly used for classification tasks such as spam detection, sentiment analysis, and document categorization.The term "naïve" reflects the conditional independence assumption, which is often unrealistic in real-world data.

ques 13  What is Bayes’ Theorem ?

Bayes’ Theorem is a fundamental concept in probability theory that describes how to update the probability of a hypothesis based on new evidence. It provides a mathematical framework for revising existing predictions or theories (updating probabilities) given new or additional evidence.

ques 14  Explain the differences between Gaussian Naïve Bayes, Multinomial Naïve Bayes, and Bernoulli Naïve Bayes .

**Gaussian Naïve Bayes**: Assumes that each feature follows a normal distribution. It's ideal when dealing with continuous data where features like age, height, or temperature are involved.


**Multinomial Naïve Bayes**: Particularly effective for text classification tasks where features represent the frequency of words or tokens. It assumes that features are generated from a multinomial distribution, making it suitable for tasks like document categorization or spam detection.


**Bernoulli Naïve Bayes**: Best suited for binary/boolean features, where each feature represents the presence or absence of a particular attribute. It's commonly used in scenarios where the mere occurrence of a feature is significant, such as detecting the presence of specific words in a document

ques 15 When should you use Gaussian Naïve Bayes over other variants?

Gaussian Naïve Bayes is a robust and efficient classifier for continuous data that follows a normal distribution. It is particularly useful for high-dimensional datasets and serves as a strong baseline model. However, careful consideration of the data distribution and feature independence is essential for optimal performance

ques 16 What are the key assumptions made by Naïve Bayes ?

The Naïve Bayes classifier is based on simplifying assumptions that make it computationally efficient and effective for certain types of data, particularly when features are conditionally independent given the class label. While these assumptions may not always hold true, the classifier often delivers robust performance in practice.

ques 17  What are the advantages and disadvantages of Naïve Bayes ?

**Advantages** :

Simplicity & Speed: Easy to implement and computationally efficient, making it suitable for large datasets.

Effective with Categorical Data: Performs well when features are categorical, such as in text classification tasks.

Handles Missing Data: Can handle missing values by ignoring instances with missing data during training.

**Disadvantages** :

Independence Assumption: Assumes that features are independent given the class label, which may not hold true in real-world data.


Zero Frequency Problem: Assigns zero probability to unseen feature values during prediction, potentially leading to issues.

Sensitive to Irrelevant Features: Performance can degrade if irrelevant features are included in the model.


ques 18 Why is Naïve Bayes a good choice for text classification ?


 Naïve Bayes is good choice for Text Classification beacuse of following :

**Simplicity & Speed**
Naïve Bayes is easy to understand and implement, making it ideal for quick prototyping and real-time applications.

**Handles High-Dimensional Data Well**
Text data often involves a vast number of features (words or tokens). Naïve Bayes effectively manages this high-dimensional space without significant performance degradation.

**Performs Well with Limited Data**
It can achieve competitive results even with relatively small amounts of training data, which is beneficial when labeled data is scarce.

**Robust to Irrelevant Features**
Due to its independence assumption, Naïve Bayes maintains robust performance even when faced with irrelevant features, which can be common in text data.

**Probabilistic Interpretation**
The model provides easily interpretable probabilities, offering insights into the likelihood of a document belonging to a particular class

ques 19 Compare SVM and Naïve Bayes for classification tasks .

**SVM (Support Vector Machine)**:

Model Type: Deterministic, geometric classifier based on hyperplanes.

Assumptions: Assumes data is separable by a hyperplane; can handle non-linear boundaries using kernel functions.



Performance: Generally high accuracy, especially with complex or high-dimensional data.


Use Cases: Effective for complex datasets with clear margins of separation.



 **Naïve Bayes** :

Model Type: Probabilistic classifier based on Bayes' theorem.

Assumptions: Assumes feature independence given the class label.


Performance: Good performance on simple or linearly separable data; may struggle with feature dependencies.


Use Cases: Ideal for applications requiring speed and scalability, such as spam filtering and sentiment analysis

ques 20 How does Laplace Smoothing help in Naïve Bayes?

Laplace Smoothing, also known as additive smoothing, is a technique used in Naïve Bayes classifiers to address the issue of zero probabilities. Without smoothing, if a feature (such as a word) does not appear in the training data for a particular class, the probability of that feature occurring in that class would be zero. This can lead to the entire product of probabilities being zero, rendering the model unable to make predictions for instances containing that feature

```python
import numpy as np
import pandas as pd
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

iris = datasets.load_iris()
```

```python
# ques 21 Write a Python program to train an SVM Classifier on the Iris dataset and evaluate accuracy



X = iris.data
y = iris.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

svm_classifier = SVC(kernel='rbf', gamma='scale')

svm_classifier.fit(X_train, y_train)
y_pred = svm_classifier.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy * 100:.2f}%')

```

```python
# ques 22 Write a Python program to train two SVM classifiers with Linear and RBF kernels on the Wine dataset, then compare their accuracies

wine = datasets.load_wine()
X = wine.data
y = wine.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

svm_linear = SVC(kernel='linear', C=1)
svm_rbf = SVC(kernel='rbf', C=1, gamma='scale')

svm_linear.fit(X_train, y_train)
svm_rbf.fit(X_train, y_train)

y_pred_linear = svm_linear.predict(X_test)
y_pred_rbf = svm_rbf.predict(X_test)
accuracy_linear = accuracy_score(y_test, y_pred_linear)
accuracy_rbf = accuracy_score(y_test, y_pred_rbf)

print(f"Accuracy of SVM with Linear Kernel: {accuracy_linear * 100:.2f}%")
print(f"Accuracy of SVM with RBF Kernel: {accuracy_rbf * 100:.2f}%")
```

```python
# ques 23 Write a Python program to train an SVM Regressor (SVR) on a housing dataset and evaluate it using Mean Squared Error (MSE)

from sklearn.datasets import fetch_california_housing
from sklearn.metrics import mean_squared_error
from sklearn.svm import SVR
from sklearn.preprocessing import StandardScaler

X, y = fetch_california_housing(return_X_y=True, as_frame=True)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
svr = SVR(kernel='rbf', C=100, gamma=0.1, epsilon=0.1)
svr.fit(X_train_scaled, y_train)
y_pred = svr.predict(X_test_scaled)
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error: {mse:.2f}")
```

```python
# ques 24 Write a Python program to train an SVM Classifier with a Polynomial Kernel and visualize the decision boundary


from matplotlib.colors import ListedColormap
import matplotlib.pyplot as plt
X, y = datasets.make_classification(n_samples=100, n_features=2, n_informative=2, n_redundant=0, n_clusters_per_class=2, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
svm_poly = SVC(kernel='poly', degree=3, coef0=1, C=1)
svm_poly.fit(X_train, y_train)
y_pred = svm_poly.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy * 100:.2f}%")
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02),
                     np.arange(y_min, y_max, 0.02))

Z = svm_poly.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.contourf(xx, yy, Z, alpha=0.75, cmap=ListedColormap(('red', 'blue')))
plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train, marker='o', edgecolors='k', cmap=ListedColormap(('darkred', 'darkblue')))
plt.scatter(X_test[:, 0], X_test[:, 1], c=y_test, marker='*', edgecolors='k', cmap=ListedColormap(('red', 'blue')))
plt.title(f"SVM with Polynomial Kernel (Degree=3)\nAccuracy: {accuracy * 100:.2f}%")
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

```python
# ques 26 Write a Python program to train a Multinomial Naïve Bayes classifier for text classification using the 20 Newsgroups dataset.

from sklearn.datasets import fetch_20newsgroups
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report

categories = ['alt.atheism', 'comp.graphics', 'sci.space', 'talk.religion.misc']
newsgroups_train = fetch_20newsgroups(subset='train', categories=categories, remove=('headers', 'footers', 'quotes'))
newsgroups_test = fetch_20newsgroups(subset='test', categories=categories, remove=('headers', 'footers', 'quotes'))

vectorizer = TfidfVectorizer(stop_words='english', max_df=0.5, min_df=5)
X_train = vectorizer.fit_transform(newsgroups_train.data)
X_test = vectorizer.transform(newsgroups_test.data)
nb_classifier = MultinomialNB()
nb_classifier.fit(X_train, newsgroups_train.target)
y_pred = nb_classifier.predict(X_test)
accuracy = accuracy_score(newsgroups_test.target, y_pred)
print(f"Accuracy: {accuracy * 100:.2f}%")
print("\nClassification Report:")
print(classification_report(newsgroups_test.target, y_pred, target_names=newsgroups_test.target_names))
```

```python
# ques 32 Write a Python program to train an SVM Classifier on an imbalanced dataset and apply class weighting and check it improve accuracy

from sklearn.datasets import make_classification
from sklearn.utils.class_weight import compute_class_weight
X, y = make_classification(
    n_samples=1000,
    n_features=20,
    n_classes=2,
    weights=[0.9, 0.1],
    flip_y=0,
    random_state=42
)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
class_weights = compute_class_weight('balanced', classes=np.unique(y), y=y_train)
class_weight_dict = dict(zip(np.unique(y), class_weights))
svm_no_weight = SVC(kernel='rbf', random_state=42)
svm_no_weight.fit(X_train, y_train)
y_pred_no_weight = svm_no_weight.predict(X_test)
accuracy_no_weight = accuracy_score(y_test, y_pred_no_weight)
svm_weighted = SVC(kernel='rbf', class_weight=class_weight_dict, random_state=42)
svm_weighted.fit(X_train, y_train)
y_pred_weighted = svm_weighted.predict(X_test)
accuracy_weighted = accuracy_score(y_test, y_pred_weighted)
print("Accuracy without class weighting:", accuracy_no_weight)
print("Accuracy with class weighting:", accuracy_weighted)
print("\nClassification Report (Weighted):\n", classification_report(y_test, y_pred_weighted))
plt.figure(figsize=(8, 6))
plt.bar(['Without Weighting', 'With Weighting'], [accuracy_no_weight, accuracy_weighted], color=['red', 'green'])
plt.title('SVM Classifier Accuracy Comparison')
plt.ylabel('Accuracy')
plt.show()
```

```python
# ques 33 Write a Python program to implement a Naïve Bayes classifier for spam detection using email data

import os
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report
import joblib


# 1. Load the dataset
# Specify the encoding, common alternatives are 'latin-1' or 'cp1252'
try:
    df = pd.read_csv('/content/spam.csv', encoding='latin-1')
except UnicodeDecodeError:
    df = pd.read_csv('/content/spam.csv', encoding='cp1252')

df.rename(columns={'v1': 'label', 'v2': 'text'}, inplace=True)
df['text'] = df['text'].apply(lambda x: x.lower())
df['label'] = df['label'].map({'spam': 1, 'ham': 0})
X_train, X_test, y_train, y_test = train_test_split(df['text'], df['label'], test_size=0.2, random_state=42)
vectorizer = TfidfVectorizer(max_features=5000)
X_train_tfidf = vectorizer.fit_transform(X_train)
X_test_tfidf = vectorizer.transform(X_test)
model = MultinomialNB()
model.fit(X_train_tfidf, y_train)
y_pred = model.predict(X_test_tfidf)
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy * 100:.2f}%')
print('Classification Report:')
print(classification_report(y_test, y_pred))
joblib.dump(model, 'spam_classifier_model.pkl')
joblib.dump(vectorizer, 'tfidf_vectorizer.pkl')
```

```python
# ques 34 Write a Python program to train an SVM Classifier and a Naïve Bayes Classifier on the same dataset and compare their accuracy

import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import SVC
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report

# 1. Load the dataset
# Assuming 'spam_data.csv' has columns 'text' and 'label'
try:
    df = pd.read_csv('/content/spam.csv', encoding='latin-1')
except UnicodeDecodeError:
    df = pd.read_csv('/content/spam.csv', encoding='cp1252')
df.rename(columns={'v1': 'label', 'v2': 'text'}, inplace=True)
df['text'] = df['text'].apply(lambda x: x.lower())
df['label'] = df['label'].map({'spam': 1, 'ham': 0})
X_train, X_test, y_train, y_test = train_test_split(df['text'], df['label'], test_size=0.2, random_state=42)
vectorizer = TfidfVectorizer(max_features=5000)
X_train_tfidf = vectorizer.fit_transform(X_train)
X_test_tfidf = vectorizer.transform(X_test)
svm_classifier = SVC(kernel='linear', random_state=42)
svm_classifier.fit(X_train_tfidf, y_train)
nb_classifier = MultinomialNB()
nb_classifier.fit(X_train_tfidf, y_train)
svm_predictions = svm_classifier.predict(X_test_tfidf)
nb_predictions = nb_classifier.predict(X_test_tfidf)
svm_accuracy = accuracy_score(y_test, svm_predictions)
nb_accuracy = accuracy_score(y_test, nb_predictions)

print(f'SVM Accuracy: {svm_accuracy * 100:.2f}%')
print(f'Naïve Bayes Accuracy: {nb_accuracy * 100:.2f}%')

print("\nSVM Classification Report:")
print(classification_report(y_test, svm_predictions))

print("\nNaïve Bayes Classification Report:")
print(classification_report(y_test, nb_predictions))
```

```python
# ques 35 Write a Python program to perform feature selection before training a Naïve Bayes classifier and compare results

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.feature_selection import SelectKBest, chi2
from sklearn.metrics import accuracy_score, classification_report


try:
    df = pd.read_csv('/content/spam.csv', encoding='latin-1')
except UnicodeDecodeError:
    df = pd.read_csv('/content/spam.csv', encoding='cp1252')
df.rename(columns={'v1': 'label', 'v2': 'text'}, inplace=True)

df['text'] = df['text'].apply(lambda x: x.lower())
df['label'] = df['label'].map({'spam': 1, 'ham': 0})
X_train, X_test, y_train, y_test = train_test_split(df['text'], df['label'], test_size=0.2, random_state=42)
vectorizer = TfidfVectorizer(max_features=5000)
X_train_tfidf = vectorizer.fit_transform(X_train)
X_test_tfidf = vectorizer.transform(X_test)
selector = SelectKBest(chi2, k=1000)
X_train_selected = selector.fit_transform(X_train_tfidf, y_train)
X_test_selected = selector.transform(X_test_tfidf)
nb_no_fs = MultinomialNB()
nb_no_fs.fit(X_train_tfidf, y_train)
y_pred_no_fs = nb_no_fs.predict(X_test_tfidf)
nb_with_fs = MultinomialNB()
nb_with_fs.fit(X_train_selected, y_train)
y_pred_with_fs = nb_with_fs.predict(X_test_selected)
accuracy_no_fs = accuracy_score(y_test, y_pred_no_fs)
accuracy_with_fs = accuracy_score(y_test, y_pred_with_fs)

print(f'Accuracy without feature selection: {accuracy_no_fs * 100:.2f}%')
print(f'Accuracy with feature selection: {accuracy_with_fs * 100:.2f}%')

print("\nClassification Report (Without Feature Selection):")
print(classification_report(y_test, y_pred_no_fs))

print("\nClassification Report (With Feature Selection):")
print(classification_report(y_test, y_pred_with_fs))
```

```python
# ques 36  Write a Python program to train an SVM Classifier using One-vs-Rest (OvR) and One-vs-One (OvO) strategies on the Wine dataset and compare their accuracy.

import numpy as np
import pandas as pd
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.multiclass import OneVsRestClassifier, OneVsOneClassifier
from sklearn.metrics import accuracy_score, classification_report
wine = datasets.load_wine()
X = wine.data
y = wine.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
svm = SVC(kernel='linear', random_state=42)
ovr_classifier = OneVsRestClassifier(svm)
ovr_classifier.fit(X_train, y_train)
y_pred_ovr = ovr_classifier.predict(X_test)
accuracy_ovr = accuracy_score(y_test, y_pred_ovr)
ovo_classifier = OneVsOneClassifier(svm)
ovo_classifier.fit(X_train, y_train)
y_pred_ovo = ovo_classifier.predict(X_test)
accuracy_ovo = accuracy_score(y_test, y_pred_ovo)
print(f"Accuracy (One-vs-Rest): {accuracy_ovr * 100:.2f}%")
print(f"Accuracy (One-vs-One): {accuracy_ovo * 100:.2f}%")

print("\nClassification Report (One-vs-Rest):")
print(classification_report(y_test, y_pred_ovr))

print("\nClassification Report (One-vs-One):")
print(classification_report(y_test, y_pred_ovo))
```

```python
# ques 38 Write a Python program to train an SVM Classifier using Stratified K-Fold Cross-Validation and compute the average accuracy

import numpy as np
from sklearn import datasets
from sklearn.model_selection import StratifiedKFold
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

data = datasets.load_breast_cancer()
X = data.data
y = data.target
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
svm = SVC(kernel='linear', random_state=42)
accuracy_scores = []
for train_index, test_index in skf.split(X, y):
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]
    svm.fit(X_train, y_train)
    y_pred = svm.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    accuracy_scores.append(accuracy)
    print(f'Accuracy for this fold: {accuracy:.2f}')
average_accuracy = np.mean(accuracy_scores)
print(f'\nAverage Accuracy: {average_accuracy:.2f}')
```

```python
# ques 39 Write a Python program to train a Naïve Bayes classifier using different prior probabilities and compare performance

import numpy as np
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score

data = datasets.load_breast_cancer()
X = data.data
y = data.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
priors_list = [
    [0.5, 0.5],
    [0.7, 0.3],
    [0.3, 0.7]
]
accuracy_scores = []
for priors in priors_list:
    nb = GaussianNB(priors=priors)
    nb.fit(X_train, y_train)
    y_pred = nb.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    accuracy_scores.append(accuracy)
    print(f"Accuracy with priors {priors}: {accuracy * 100:.2f}%")
best_accuracy = max(accuracy_scores)
best_priors = priors_list[accuracy_scores.index(best_accuracy)]
print(f"\nBest accuracy achieved with priors {best_priors}: {best_accuracy * 100:.2f}%")
```

```python
# ques 43 Write a Python program to train an SVM Classifier and visualize the Confusion Matrix using seaborn

import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix

data = datasets.load_breast_cancer()
X = data.data
y = data.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
svm = SVC(kernel='linear', random_state=42)
svm.fit(X_train, y_train)
y_pred = svm.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=data.target_names, yticklabels=data.target_names)
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Confusion Matrix')
plt.show()
```

```python
# ques 44 Write a Python program to train an SVM Regressor (SVR) and evaluate its performance using Mean Absolute Error (MAE) instead of MSE

import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.svm import SVR
from sklearn.metrics import mean_absolute_error
from sklearn.preprocessing import StandardScaler

data = datasets.fetch_california_housing()
X = data.data
y = data.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
svr = SVR(kernel='rbf', C=1.0, epsilon=0.1)
svr.fit(X_train_scaled, y_train)
y_pred = svr.predict(X_test_scaled)
mae = mean_absolute_error(y_test, y_pred)
print(f"Mean Absolute Error: {mae:.4f}")
```

```python
# ques 45  Write a Python program to train a Naïve Bayes classifier and evaluate its performance using the ROC-AUC score

import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import roc_curve, auc

data = datasets.load_breast_cancer()
X = data.data
y = data.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
nb = GaussianNB()
nb.fit(X_train, y_train)
y_pred_proba = nb.predict_proba(X_test)[:, 1]
fpr, tpr, _ = roc_curve(y_test, y_pred_proba)
roc_auc = auc(fpr, tpr)
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (area = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc='lower right')
plt.show()
```

```python
# ques 46 Write a Python program to train an SVM Classifier and visualize the Precision-Recall Curve.

import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import precision_recall_curve, average_precision_score

data = datasets.load_breast_cancer()
X = data.data
y = data.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
svm = SVC(kernel='linear', probability=True, random_state=42)
svm.fit(X_train, y_train)
y_pred_proba = svm.predict_proba(X_test)[:, 1]
precision, recall, _ = precision_recall_curve(y_test, y_pred_proba)
avg_precision = average_precision_score(y_test, y_pred_proba)
plt.figure(figsize=(8, 6))
plt.plot(recall, precision, color='b', alpha=0.2, label=f'Precision-Recall curve (AP = {avg_precision:.2f})')
plt.fill_between(recall, precision, alpha=0.2, color='b')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('2-class Precision-Recall curve')
plt.legend(loc='lower left')
plt.show()
```
