ques 1 What is a random variable in probability theory?

 A random variable is a fundamental concept in probability theory. It's essentially a variable that takes on different values based on the outcome of a random experiment or process. ques 3 What is the difference between discrete and continuous distributions?

Differences between discrete and continuous distribution :

**1. Nature of Random Variable:**

Discrete Distribution: Describes random variables that can only take specific, distinct values. These values are countable (either finite or countably infinite).

EXAMPLE :  Number of heads in 10 coin flips, number of cars passing a checkpoint in an hour, number of students in a class.

Discrete Distribution: The probability function is a probability mass function (PMF), which gives the probability of each distinct outcome.

3. **Probability of Single Points:**

Discrete Distribution: The probability of a random variable taking a particular value is non-zero.

Continuous Distribution: The probability of the random variable taking exactly any particular value is always zero. Probabilities are assigned to intervals or ranges.

ques 16  How do you calculate a Z-score, and what does it represent?

A Z-score (also called a standard score or Z-value) measures how many standard deviations a specific data point (or observation) is away from the mean of the distribution. It helps to standardize data, making it easier to compare values from different distributions.

Steps for Calculating a Z-Score:

1. Identify the value X: This is the specific data point or observation you're interested in.

2. Determine the mean ùúáŒº: This is the average of the dataset or population from which X is drawn.

3. Calculate the standard deviation œÉ: The standard deviation tells you how spread out the data is around the mean.

4. Plug values into the Z-score formula: Subtract the mean from the value X, and then divide by the standard deviation.

The Z-score represents the number of standard deviations a data point is away from the mean. It is used for comparing individual data points to the mean of a distribution, identifying outliers, standardizing data, and calculating probabilities. A Z-score allows you to understand how extreme or typical a value is within the context of the data set.

```python
# ques 3 Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution

def bernoulli_pdf(x, p):
    """
    Calculate the Probability Distribution Function (PDF) of a Bernoulli distribution.

    Parameters:
    x (int): The outcome (0 or 1).
    p (float): The probability of success (0 <= p <= 1).

    Returns:
    float: The probability of the outcome x.
    """
    if x not in [0, 1]:
        raise ValueError("x must be either 0 or 1")

    if p < 0 or p > 1:
        raise ValueError("p must be between 0 and 1 inclusive")

    if x == 1:
        return p
    else:
        return 1 - p
p = 0.7
x1 = 1
x0 = 0
print(f"P(X = 1) = {bernoulli_pdf(x1, p)}")
print(f"P(X = 0) = {bernoulli_pdf(x0, p)}")
```

```python
# ques 9 Write a Python function to calculate Z-scores from a dataset and plot them


def calculate_z_scores(data):
    """
    Calculate the Z-scores for a given dataset.

    Parameters:
    data (array-like): The dataset from which to calculate Z-scores.

    Returns:
    array: The Z-scores for each element in the dataset.
    """

    mean = np.mean(data)
    std_dev = np.std(data)
    z_scores = (data - mean) / std_dev

    return z_scores
data = np.random.normal(0, 1, 1000)
z_scores = calculate_z_scores(data)
plt.figure(figsize=(8, 6))
plt.hist(z_scores, bins=30, edgecolor='black', color='skyblue', alpha=0.7, density=True)
plt.title("Histogram of Z-scores")
plt.xlabel("Z-score")
plt.ylabel("Density")
plt.grid(True)
plt.show()
```

```python
# ques 10 Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution.


def central_limit_theorem(n_samples, sample_size, distribution, *params):
    """
    Demonstrates the Central Limit Theorem using the specified distribution.

    Parameters:
    - n_samples: Number of random samples to draw
    - sample_size: Size of each sample
    - distribution: The distribution to sample from (e.g., np.random.exponential)
    - *params: Additional parameters for the distribution (like scale for exponential)

    Returns:
    - sample_means: List of means of the samples
    """
    sample_means = []
    for _ in range(n_samples):
        sample = distribution(*params, size=sample_size)
        sample_mean = np.mean(sample)
        sample_means.append(sample_mean)

    return sample_means
n_samples = 1000
sample_size = 50
distribution = np.random.exponential
lambda_param = 1
sample_means = central_limit_theorem(n_samples, sample_size, distribution, lambda_param)
plt.hist(sample_means, bins=30, edgecolor='black', alpha=0.7, color='skyblue', density=True)
plt.title(f"Central Limit Theorem - Sample Means from Exponential Distribution\nSample Size={sample_size}, n_samples={n_samples}")
plt.xlabel("Sample Mean")
plt.ylabel("Density")
plt.grid(True)
plt.show()
```

```python
# ques 11 Simulate multiple samples from a normal distribution and verify the Central Limit Theorem

import numpy as np
import matplotlib.pyplot as plt

def central_limit_theorem_normal(n_samples, sample_size, mean, std_dev):
    """
    Demonstrates the Central Limit Theorem using a normal distribution.

    Parameters:
    - n_samples: Number of samples to draw
    - sample_size: Size of each sample
    - mean: Mean of the normal distribution
    - std_dev: Standard deviation of the normal distribution

    Returns:
    - sample_means: List of means of the samples
    """
    sample_means = []
    for _ in range(n_samples):
        sample = np.random.normal(mean, std_dev, sample_size)
        sample_mean = np.mean(sample)
        sample_means.append(sample_mean)

    return sample_means
mean = 50
std_dev = 10
n_samples = 1000
sample_size = 50
sample_means = central_limit_theorem_normal(n_samples, sample_size, mean, std_dev)
plt.hist(sample_means, bins=30, edgecolor='black', alpha=0.7, color='skyblue', density=True)
plt.title(f"Central Limit Theorem - Sample Means from Normal Distribution\nSample Size={sample_size}, n_samples={n_samples}")
plt.xlabel("Sample Mean")
plt.ylabel("Density")
plt.grid(True)
plt.show()
```

```python
# ques 13  Generate random variables and calculate their corresponding probabilities using the binomial distribution

from scipy.stats import binom

def binomial_distribution(n, p, size):
    """
    Generate random variables from a binomial distribution and calculate the probabilities.

    Parameters:
    - n: Number of trials
    - p: Probability of success on each trial
    - size: Number of random variables to generate

    Returns:
    - random_vars: Random variables generated from the binomial distribution
    - probabilities: Corresponding probabilities for each outcome
    """
    random_vars = np.random.binomial(n, p, size)
    k_values = np.arange(0, n+1)
    probabilities = binom.pmf(k_values, n, p)

    return random_vars, probabilities, k_values
n = 10
p = 0.5
size = 1000
random_vars, probabilities, k_values = binomial_distribution(n, p, size)
plt.hist(random_vars, bins=np.arange(0, n+2) - 0.5, edgecolor='black', density=True, alpha=0.7, color='skyblue', label="Generated Data")
plt.plot(k_values, probabilities, 'o', color='red', label="Binomial PMF")
plt.title(f"Binomial Distribution (n={n}, p={p})")
plt.xlabel("Number of Successes (k)")
plt.ylabel("Probability")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#ques 14 Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution



def calculate_z_score(x, mean, std_dev):
    """
    Calculate the Z-score for a given data point.

    Parameters:
    - x: The data point
    - mean: The mean of the dataset
    - std_dev: The standard deviation of the dataset

    Returns:
    - z_score: The calculated Z-score
    """
    z_score = (x - mean) / std_dev
    return z_score

def plot_standard_normal_distribution(z_score):
    """
    Plot the standard normal distribution and indicate the position of the Z-score.

    Parameters:
    - z_score: The Z-score to plot on the distribution
    """
    x = np.linspace(-4, 4, 1000)
    pdf = norm.pdf(x, loc=0, scale=1)
    plt.plot(x, pdf, label="Standard Normal Distribution", color='blue', linewidth=2)
    plt.axvline(z_score, color='red', linestyle='dashed', label=f"Z-score = {z_score:.2f}")
    plt.title(f"Standard Normal Distribution and Z-score = {z_score:.2f}")
    plt.xlabel("Value")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid(True)
    plt.show()
mean = 50
std_dev = 10
x = 60
z_score = calculate_z_score(x, mean, std_dev)
print(f"Z-score for x = {x}: {z_score:.2f}")
plot_standard_normal_distribution(z_score)
probability = norm.cdf(z_score)
print(f"The cumulative probability (CDF) for Z-score = {z_score:.2f} is {probability:.4f}")
```

```python
#ques 15 Implement hypothesis testing using Z-statistics for a sample dataset

import scipy.stats as stats
def z_test(sample_data, population_mean, population_std, sample_size, alpha=0.05):
    """
    Perform a Z-test to determine if the sample mean is significantly different from the population mean.

    Parameters:
    - sample_data: Sample data (array of values)
    - population_mean: Population mean (H0)
    - population_std: Population standard deviation
    - sample_size: Sample size (n)
    - alpha: Significance level (default = 0.05)

    Returns:
    - z_statistic: Calculated Z-statistic
    - p_value: Corresponding p-value
    - decision: Reject or fail to reject the null hypothesis
    """
    sample_mean = np.mean(sample_data)
    z_statistic = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

    p_value = 2 * (1 - stats.norm.cdf(abs(z_statistic)))
    if p_value <= alpha:
        decision = "Reject the null hypothesis (There is a significant difference)"
    else:
        decision = "Fail to reject the null hypothesis (No significant difference)"

    return z_statistic, p_value, decision
np.random.seed(42)
sample_data = np.random.normal(52, 10, 30)
population_mean = 50
population_std = 10
sample_size = len(sample_data)
z_statistic, p_value, decision = z_test(sample_data, population_mean, population_std, sample_size)
print(f"Z-statistic: {z_statistic:.4f}")
print(f"P-value: {p_value:.4f}")
print(f"Decision: {decision}")
```

```python
# ques 16 Create a confidence interval for a dataset using Python and interpret the result

import numpy as np
import scipy.stats as stats

def confidence_interval(data, confidence_level=0.95):
    """
    Calculate the confidence interval for the mean of the dataset.

    Parameters:
    - data: The sample dataset (array of values)
    - confidence_level: The confidence level (default = 0.95)

    Returns:
    - lower_bound: Lower bound of the confidence interval
    - upper_bound: Upper bound of the confidence interval
    """
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)
    n = len(data)
    sem = sample_std / np.sqrt(n)
    if n > 30:
        critical_value = stats.norm.ppf(1 - (1 - confidence_level) / 2)
    else:
        critical_value = stats.t.ppf(1 - (1 - confidence_level) / 2, df=n-1)
    margin_of_error = critical_value * sem
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound, sample_mean
np.random.seed(42)
sample_data = np.random.normal(100, 15, 50)
confidence_level = 0.95
lower_bound, upper_bound, sample_mean = confidence_interval(sample_data, confidence_level)

print(f"Sample Mean: {sample_mean:.2f}")
print(f"{int(confidence_level * 100)}% Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")
print("\nInterpretation:")
print(f"The true population mean is estimated to be between {lower_bound:.2f} and {upper_bound:.2f}")
print(f"We are {confidence_level * 100}% confident that the true population mean lies within this range.")
```

```python
#ques 20  Simulate a random variable using a continuous uniform distribution and calculate its expected value

def simulate_uniform_distribution(a, b, num_samples):
    samples = np.random.uniform(a, b, num_samples)
    expected_value_theoretical = (a + b) / 2
    sample_mean = np.mean(samples)
    return samples, expected_value_theoretical, sample_mean
a = 2
b = 8
num_samples = 10000
samples, expected_value_theoretical, sample_mean = simulate_uniform_distribution(a, b, num_samples)
print(f"Theoretical Expected Value: {expected_value_theoretical:.2f}")
print(f"Empirical Sample Mean (approx. Expected Value): {sample_mean:.2f}")
```

```python
#ques 23  Implement Z-score normalization on a dataset and visualize its transformation

np.random.seed(42)
data = np.random.normal(loc=50, scale=10, size=1000)
mean = np.mean(data)
std_dev = np.std(data)
z_scores = (data - mean) / std_dev
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.hist(data, bins=30, color='orange', edgecolor='black')
plt.title('Original Data (Before Z-score Normalization)')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.subplot(1, 2, 2)
plt.hist(z_scores, bins=30, color='purple', edgecolor='black')
plt.title('Normalized Data (After Z-score Normalization)')
plt.xlabel('Z-score')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
print(f"Mean of normalized data: {np.mean(z_scores):.2f}")
print(f"Standard Deviation of normalized data: {np.std(z_scores):.2f}")
```

```python
#ques 24  Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution.

import numpy as np
from scipy.stats import skew, kurtosis

def calculate_skewness_kurtosis(data):
    skewness = skew(data)
    kurt = kurtosis(data)
    return skewness, kurt
np.random.seed(42)
mean = 50
std_dev = 10
size = 1000
data = np.random.normal(loc=mean, scale=std_dev, size=size)
skewness, kurt = calculate_skewness_kurtosis(data)
print(f"Skewness of the dataset: {skewness:.4f}")
print(f"Kurtosis of the dataset: {kurt:.4f}")
```
